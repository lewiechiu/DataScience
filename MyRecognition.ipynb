```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Acquire data from Taipei Metro system
This notebook demos how I can use pandas to read data from html. Then do some data preparation to implement visualization.

```python
url = "https://web.metro.taipei/RidershipCounts/c/10701.htm"
data = pd.read_html(url)
cut_off = (len(data[0])-4)


for i in range(1,5):
    url = list(url)
    url[-5] = str(i)
    url = ''.join(url)
    
    data = pd.read_html(url)
    cut_off = (len(data[0])-4)
    data = np.array(data[0][2:cut_off][2], dtype = 'int')

    days = np.arange(1,cut_off-1,1)
    a = plt.plot(days,data)
```
