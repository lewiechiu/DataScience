# DUO

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/train_V2.csv")
```

```python
mapings = lambda x: 'solo' if (x == 'solo' or x == 'solo-fpp') else 'duo' if (x == 'duo') or (x == 'duo-fpp') else 'squad' if (x == 'squad') or (x == 'squad-fpp') else 'others'
df['matchType'] = df['matchType'].apply(mapings)
```

```python
df = df.drop( df[df['matchType'] == 'others'].index )
df = df.drop( df[df['matchType'] == 'solo'].index )
df = df.drop( df[df['matchType'] == 'squad'].index )
df.head()
```

```python
df.columns
```

### DUO 
How many people are there in the duo??

```python
duo_df = df
matches = duo_df.groupby(['matchId', 'groupId'])
cnt = 0
player_cnt=[1] * 80
player_cnt
for _ ,i in matches:
    player_cnt[len(i.Id)] += 1
```

```python
plt.plot(np.log10(player_cnt))
plt.title("Distribution of players")
plt.ylabel("counts in log_10")
plt.show()
```

From the distribution, we see most of the game of match-type *DUO*, have players less than 10.

For our convenience, we want to select those datas of which groups has size 2, and both of them has $winPlacePerc=1$.

```python
for i in duo_df.columns:
    if i != 'matchId' and i != 'winPlacePerc' and i!= 'matchType' and i != 'Id' and i != 'groupId':
        duo_df[i] =  duo_df.groupby('matchId')[i].rank(pct=True, axis=1)
```

```python
duo_df.head(20)
```

```python
matches = duo_df.groupby(['matchId', 'groupId'])
matches = matches.filter(lambda x: (len(x.Id) == 2) and (sum(x.winPlacePerc) > 1.9) )

matches.head()
```

```python
len(matches.values), len(set(matches.matchId))
```

Now the matches variable is what we want for selecting only matches restricting groups to be of size 2.

```python
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D
pca = PCA(n_components=3)

matches_pca = matches.drop(['matchId', 'winPlacePerc', 'groupId', 'Id'], axis=1)
matches_pca = matches_pca.drop(['matchType'], axis = 1)
winner = pca.fit_transform(matches_pca)

x = []
y = []
z = []
fig = plt.figure(figsize=(15,11))
ax = Axes3D(fig)

for i in winner:
    x.append(i[0])
    y.append(i[1])
    z.append(i[2])
print(len(x))
ax.scatter(x,y,z, s=0.4)
ax.view_init(40,10)
plt.title("distibution of PCA winner")
plt.show()
```

From the above 3d scatter plot, we found out that it kind of indicates that we have two very disticnt type of players in the game. In SOLO, we have already used SOTA techniques to prove that there does exist clustering in higher dimension space. What can be asked further is that , what does all these points mean?
